```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

## Read in Data
df = pd.read_csv("all_data.csv")
df = df.rename({'Life expectancy at birth (years)': 'Life_expectancy'}, axis=1)
# df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1)
print(df.head())
countries = df.Country.unique()
print(countries)
```

First look at line charts 

```python
x = df.Year.unique()

for country in countries:
    plt.plot(x, df.GDP[df.Country==country])

plt.legend(countries)
plt.show()
plt.clf()

for country in countries:
    plt.plot(x, df.Life_expectancy[df.Country==country])

plt.legend(countries)
plt.show()
plt.clf()
```
